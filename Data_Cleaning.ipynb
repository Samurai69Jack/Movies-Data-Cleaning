```python
import pandas as pd

df=pd.read_csv('movies.csv')
df
```

```python
df=df.drop_duplicates()
df
```

```python
df=df.drop(columns=['Gross'])
df
```

```python
df2=df.copy()
df2.columns=[x.lower() for x in df2.columns]
df2
```

```python
df2.columns=df2.columns.str.replace("-","_")
df2
```

```python
df3=df2.copy()
df3
```

```python
df3['year']=df3['year'].str.extract(r"(\d{4})")
df3
```

```python
df3['year']=df3['year'].astype(float).astype("Int64")
df3
```

```python
df4=df3.copy()
df4
```

```python
df4['rating']=df4['rating'].fillna(df4['rating'].median())
df4
```

```python
df4['runtime']=df4['runtime'].fillna(df4['runtime'].median())
df4
```

```python
df4['votes']=df4['votes'].str.replace(",","",regex=False)
df4
```

```python
df4['votes']=df4['votes'].astype(float)
df4
```

```python
df4['votes']=df4['votes'].fillna(df4['votes'].median())
df4
```
